```python
# Importando as libs
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
oportunidade = input("Qual vaga você deseja? ")
```

```python
oportunidade_tratada = oportunidade.replace(" ","%20").lower()
```

```python
oportunidade_tratada
```

```python
url = f'''https://www.linkedin.com/jobs/search?
keywords={oportunidade_tratada}
&location=Brasil
&geoId=106057199
&f_TPR=r86400&position=1
&pageNum=0'''
```

```python
url = url.replace("\n","")
url
```

```python
response = requests.get(url)
response
```

```python
response.text
```

```python
site = BeautifulSoup(response.text,"html.parser")
```

```python
site
```

```python
dados = site.find_all("div", attrs={"class" : "base-search-card__info"})
#dados
```

```python
armazenamento_vaga = []
armazenamento_empresa = []
armazenamento_localizacao = []
armazenamento_tempo = []
armazenamento_data = []


for palavra in dados:
  #Nome da vaga
  nome_vaga = palavra.find_all("h3")[0].text.strip()
  armazenamento_vaga.append(nome_vaga)

  #Nome da empresa
  nome_empresa = palavra.find_all("h4")[0].text.strip()
  armazenamento_empresa.append(nome_empresa)

  #Localização
  localizacao = palavra.find_all("span", attrs = {"class": "job-search-card__location"})[0].text.strip()
  armazenamento_localizacao.append(localizacao)

  #Tempo disponibilidade da vaga
  tempo_vaga = palavra.find_all("time")[0].text.strip()
  armazenamento_tempo.append(tempo_vaga)

  #Data
  data = palavra.find_all("time")[0]
  data_trat = data["datetime"]
  armazenamento_data.append(data_trat)

  # se quiser em uma linha só: data_trat = palavra.find_all("time")[0].get('datetime')
  # ou data_trat = palavra.find_all("time")[0]["datetime"]


```

```python
base_vagas = {
"Vaga" : armazenamento_vaga,
"Empresa": armazenamento_empresa,
"Localizacao" : armazenamento_localizacao,
"Tempo_postada": armazenamento_tempo,
"Data_vaga" : armazenamento_data
}
```

```python
pd.DataFrame(base_vagas)
```

```python
df = pd.DataFrame(base_vagas)
```

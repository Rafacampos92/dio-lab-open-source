```python
# Importando as libs
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
oportunidade = input("Qual vaga você deseja? ")
```

```python
oportunidade_tratada = oportunidade.replace(" ","%20").lower()
```

```python
oportunidade_tratada
```

```python
url = f'''https://www.linkedin.com/jobs/search?
keywords={oportunidade_tratada}
&location=Brasil
&geoId=106057199
&f_TPR=r86400&position=1
&pageNum=0'''
```

```python
url = url.replace("\n","")
url
```

```python
response = requests.get(url)
response
```
